```python
### remove columns that are not in london boroughs 

# first we need a list of boroughs - this is obtained from https://en.wikipedia.org/wiki/London_boroughs
# parse the file to obtain the list of valid boroughs 
borough_fname = 'london_boroughs.txt'
valid_boroughs = set() 
with open(borough_fname, 'rt') as f:
    for line in f: 
        line = line.strip() 
        if len(line) > 0:
            valid_boroughs.add(line.split('\t')[0].strip())

# need to remove '[note x]' from some boroughs and change 'and' to '&'
for borough in list(valid_boroughs): 
    s = borough[:borough.index('[')] if '[' in borough else borough
    valid_boroughs.remove(borough)
    valid_boroughs.add(s.replace(' and ', ' & '))
    
# now these are the columns we need to keep 
df = df.drop(columns=[col for col in df.columns if col not in valid_boroughs])
print(df.shape)
```

```python
# function to carry out the suggested calculation
def compute_price_ratio(df, year1, year2):
    """
    Function to compute the ratio of average housing prices between year1 and year2.  
    """
    # we'll do the suggested analysis for the last two decades: 2002 to 2022
    df_by_yr = df.resample('Y').mean()
    df_by_yr = df_by_yr.set_index(df_by_yr.index.year).transpose()
             
    df_yr1_yr2 = pd.DataFrame(df_by_yr[[year1, year2]])
    df_yr1_yr2['price_ratio'] = df_yr1_yr2[year2] / df_yr1_yr2[year1]

    df_yr1_yr2 = df_yr1_yr2.sort_values('price_ratio', ascending=False)
    return df_yr1_yr2
```

```python
# compute the ratio of average housing price over the last two decades: from 2002 to 2022
year1 = 2002
year2 = 2022
df_02_22 = compute_price_ratio(df, year1, year2)

# plot the housing price trend for the top k boroughs
k = 5
top_boroughs = df_02_22.index[:k]
df[top_boroughs].loc[f'{year1}':f'{year2}'].plot(ylabel='Price (millions of pounds)',
                      title=f'Average housing price per borough ({year1}-{year2})')
plt.show()
```
